```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
files.upload()
```

```python
pop=pd.read_csv("global_population_stats_2024.csv")
pop
```

```python
print("shape:",pop.shape)
print("\nColumns:",pop.columns.tolist())
print("\nData Types:\n",pop.dtypes)
print("\nFirst 5 Rows:\n",pop.head())
```

```python
pop.isnull().sum()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(pop.corr(numeric_only=True),annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
num_cols=pop.select_dtypes(include=np.number).columns

for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.histplot(pop[col],kde=True,color='skyblue')
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

In india there is the most elder peoples are living according to this report.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=pop,x="Population Aged 0 to 14 (%)", y="Population Aged 60 and Over (%)", hue="Sex ratio (males per 100 females)", palette="coolwarm", s=100)
plt.title("Age Structure vs Sex Ratio")
plt.show()
```

```python
top_pop=pop.sort_values("Population(in millions)",ascending=False).head(10)
plt.figure(figsize=(8,4))
sns.barplot(data=top_pop,x="Country",y="Population(in millions)",palette="crest")
plt.title("Top 10 Most Populous Countries")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=pop.sort_values("Sex ratio (males per 100 females)", ascending=False), x="Country", y="Sex ratio (males per 100 females)", palette="mako")
plt.title("Sex Ratio by Country")
plt.xticks(rotation=45)
plt.show()
```
